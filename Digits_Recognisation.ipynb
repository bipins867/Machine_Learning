```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
```

```python
data=load_digits()
print(data.keys())
```

```python
data.data[0]
```

```python
data.target[0]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.matshow(data.images[0])
```

```python
for i in range(5,10):
    print(data.target[i])
    plt.matshow(data.images[i])
```

```python
#Splitting the model using train_test_split
x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.25,random_state=10)
```

```python
x_train.shape
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
#Here our model percentage is approx 96%
model.predict(x_test[50:55])
```

```python
y_test[50:55]
```

```python
#Now drawing confusion Matrix
from sklearn.metrics import confusion_matrix
```

```python
ypredict=model.predict(x_test)
```

```python
cm=confusion_matrix(y_test,ypredict)
```

```python
print(cm)
```
